<h1>Solution for Problem 01</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

# First point
x1 = 0
y1 = 55

# Second point
x2 = 2
y2 = 76

# Develop slope-interception equation
m = (y2 - y1) / (x2 - x1)
b = y1 - (m * x1)
print(f"y = ({m} * x) + {b}")

# Define x-axis and y-xis min and max values
x_min = 0
x_max = 10
y_min = 0
y_max = 150

# Create a line
y3 = m * x_min + b
y4 = m * x_max + b

# Graph window, x-axis (blue), y-axis (red)
fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])
plt.plot([x_min, x_max], [0, 0], "b")
plt.plot([0, 0], [y_min, y_max], "r")

# Graph details
ax.set_xlabel("Time")
ax.set_ylabel("Population")
ax.grid(True)
ax.set_xticks(np.arange(x_min, x_max, 1))
ax.set_yticks(np.arange(y_min, y_max, 10))

# Plot linear function as green line
plt.plot([x_min, x_max], [y3, y4], "g")

plt.show()
```
