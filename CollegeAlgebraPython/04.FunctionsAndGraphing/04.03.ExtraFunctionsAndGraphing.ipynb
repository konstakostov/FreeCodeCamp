<h1>Loop Through Function Inputs</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 10

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Plot multiple points
for x in range(0, 10 + 1):
    y =  0 + x
    plt.plot([x], [y], 'go')

plt.show()
```

<h1>Use and Array as Input</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

x_min = -10
x_max = 10
y_min = -10
y_max = 10

# Define points in the array
points = 2 * (x_max - x_min)

# Array of x values
# np.linspace is a function that sets up the array
x = np.linspace(x_min, x_max, points)

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Define y-value
y = 2 * x + 1

# Plot the Graph
plt.plot(x, y, "g")

plt.show()
```

<h1>Ways to Customize a Graph</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

x_min = -10
x_max = 10
y_min = -10
y_max = 10

# Define points in the array
points = 2 * (x_max - x_min)

# Array of x values
# np.linspace is a function that sets up the array
x = np.linspace(x_min, x_max, points)

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Define x-label
ax.set_xlabel("x Values")

# Define y-label
ax.set_ylabel("y Values")

# Define title
ax.set_title("Graph")

# Show grid in graph
ax.grid(True)

# Setting tick marks on each axis of the graph
ax.set_xticks(np.arange(x_min, x_max, 1))
ax.set_yticks(np.arange(y_min, y_max, 1))

# Define y-value
y = 2 * x + 1

# Plotting multiple graphs
plt.plot(x, y, label="y = 2x + 1")
plt.plot([4], [6], "go", label="point")
plt.plot(x, 3 * x, label="Steeper Line")

# Show graph legend
plt.legend()

plt.show()
```
