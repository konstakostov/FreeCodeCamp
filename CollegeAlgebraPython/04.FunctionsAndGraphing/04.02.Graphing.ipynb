<h1>Define Dimensions of a Graph</h1>

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Add Dimensions
plt.axis([-10, 10, -10, 10])

plt.show()
```

<h1>Better Way of Setting Dimensions of a Graph</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 20

fig, ax = plt.subplots()

plt.axis([x_min, x_max, y_min, y_max])

plt.show()
```

<h1>Display Axis Lines</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 10

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

plt.show()
```

<h1>Plot 1 (One) Point</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 10

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Plot 1 point
plt.plot([5], [4], 'go')

plt.show()
```

<h1>Plot Multiple Points</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 10

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Plot multiple points
for x in range(0, 10 + 1):
    y =  0 + x
    plt.plot([x], [y], 'go')

plt.show()
```

<h1>Graph and Table of (x, y) Values</h1>

```python
import matplotlib.pyplot as plt

x_min = -10
x_max = 10
y_min = -10
y_max = 10

fig, ax = plt.subplots()

# Window size
plt.axis([x_min, x_max, y_min, y_max])

# X-axis (Blue color)
plt.plot([x_min, x_max], [0, 0], 'b')

# Y-axis (Red color)
plt.plot([0, 0], [y_min, y_max], 'r')

# Print table with (x, y) values
print("x \t y")

# Plot multiple points
for x in range(x_min, x_max + 1):
    y =  0 + x
    
    # Plot point on the graph
    plt.plot([x], [y], 'go')
    
    # Print the value in the table
    print(x, "\t",  y)

plt.show()
```
